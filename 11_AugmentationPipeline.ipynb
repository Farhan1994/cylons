```
import Augmentor
pipes = Augmentor.Pipeline('myimgs') # no need to worry about expand_dims or reshaping 
```

```
pipes.flip_left_right(0.5)
pipes.black_and_white(0.1)
pipes.random_distortion(1, 10, 10, 1)
pipes.rotate(0.5, 25, 25)
pipes.skew(0.25, 0.5)

pipes.sample(1)
```
